```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import numpy as np
import pandas as pd

```

```python
url = 'https://data.cityofnewyork.us/resource/nwxe-4ae8.json'
trees = pd.read_json(url)
trees.head(10)
```

```python
boro = 'Bronx'
BX_soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Bronx\'' +\
        '&$group=health').replace(' ', '%20')
BX_soql_url = pd.read_json(BX_soql_url)
BX_soql_url

boro = 'Brooklyn'
BK_soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Brooklyn\'' +\
        '&$group=health').replace(' ', '%20')
BK_soql_url = pd.read_json(BK_soql_url)
BK_soql_url

boro = 'Staten Island'
SI_soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Staten Island\'' +\
        '&$group=health').replace(' ', '%20')
SI_soql_url = pd.read_json(SI_soql_url)
SI_soql_url

boro = 'Manhattan'
M_soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Manhattan\'' +\
        '&$group=health').replace(' ', '%20')
M_soql_url = pd.read_json(M_soql_url)
M_soql_url


boro = 'Queens'
QN_soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Queens\'' +\
        '&$group=health').replace(' ', '%20')
QN_soql_url = pd.read_json(QN_soql_url)
QN_soql_url
```

```python
boro = 'NYC'
all_soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=spc_common,health,steward,count(tree_id)' +\
        '&$group=spc_common,steward, health').replace(' ', '%20')
all_soql_url = pd.read_json(all_soql_url)
```

```python
good = all_soql_url[all_soql_url['health'] == 'Good']
fair = all_soql_url[all_soql_url['health'] == 'Fair']
poor = all_soql_url[all_soql_url['health'] == 'Poor']
```

```python
app = dash.Dash()
server = app.server

colors = {'background':'D3D3D3', 'text':'#4a4949'}

app.layout = html.Div(children=[
    html.H1('NYC Proportion of Trees by Borough',style={'textAlign':'center',
                                                       'color':colors['text']}),
    html.Div('Tree Health',style={'textAlign':'center',
                                 'color':colors['text']}),
    dcc.Graph(id='Q1',
             figure={
            'data': [
                {'x': BX_soql_url['health'], 'y':BX_soql_url['count_tree_id'], 'type': 'bar', 'name': 'Bronx'},
                {'x': BK_soql_url['health'], 'y':BK_soql_url['count_tree_id'], 'type': 'bar', 'name': u'Brooklyn'},
                {'x': QN_soql_url['health'], 'y':QN_soql_url['count_tree_id'], 'type': 'bar', 'name': u'Queens'},
                {'x': M_soql_url['health'], 'y':M_soql_url['count_tree_id'], 'type': 'bar', 'name': u'Manhattan'},
                {'x': SI_soql_url['health'], 'y':SI_soql_url['count_tree_id'], 'type': 'bar', 'name': u'Staten Island'}
            ],
            'layout': {
                'title': 'Q1: Proportion of trees according to their health.',
                'plot_bgcolor':colors['background'],
                'paper_bgcolor':colors['background'],
                'font':{'color':colors['text']}
                
            }
        }
    ),dcc.Graph(id='Q2',
             figure={
            'data': [       
                {'x': good['steward'], 'y':good['count_tree_id'], 'type': 'bar', 'name': 'Good'},
                {'x': fair['steward'], 'y':fair['count_tree_id'], 'type': 'bar', 'name': 'Fair'},
                {'x': poor['steward'], 'y':poor['count_tree_id'], 'type': 'bar', 'name': 'Poor'},

            ],
            'layout': {
                'title': 'Q2: Are stewards having an impact on health of trees?',
                'plot_bgcolor':colors['background'],
                'paper_bgcolor':colors['background'],
                'font':{'color':colors['text'],
                'barmode':'bar'}
                
            }
        }
    )
])


if __name__ == '__main__':
    app.run_server()
```
